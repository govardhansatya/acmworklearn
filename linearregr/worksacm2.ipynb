```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
df=pd.read_csv("D:\\acm\\traildatsets\\accessories.csv")
print(df)
```

```python
df.head()
df.dropna(thresh=1).reset_index(drop=True)
```

```python
y=df['Sell']
df1=df.drop(columns='Sell')
X=df1[['Name','Variation','DIY','Buy','Color 1','Color 2','Size','Source','Mannequin Piece','Version','Style','Label Themes','Type','Villager Equippable','Catalog']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
object_cols = [col for col in X_train.columns if X_train[col].dtype == "object"]
object_cols
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
```

```python
X_train[object_cols] = ordinal_encoder.fit_transform(X_train[object_cols])
X_test[object_cols] = ordinal_encoder.transform(X_test[object_cols])
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print('variance score: {}'.format(reg.score(X_test,y_test)))
```

```python
y_predict=reg.predict(X_test)
mae=mean_absolute_error(y_predict,y_test)
```

```python
print("Mean absolute error is:- ",mae)
```

```python
r2_score(y_test,y_predict)
```
