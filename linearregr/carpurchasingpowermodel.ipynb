```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
y=df['Car Purchase Amount']
df1=df.drop(columns='Car Purchase Amount')
X=df[['Customer Name','Gender', 'Age', 'Annual Salary', 'Credit Card Debt', 'Net Worth']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.32,random_state=1)
```

```python
colcat=[colu for colu in X.columns if X[colu].dtype=='object']
colnum=[colu for colu in X.columns if X[colu].dtype!='object']
print('categorical cols', colcat)
print('numerical cols',colnum)
```

```python
df[colcat].nunique()
```

```python
df[colnum].nunique()
```

```python
scaler1=MinMaxScaler(feature_range=(0,1))
x=X
x[colnum]=scaler1.fit_transform(x[colnum])
x.head()
```

```python
object_cols = [col for col in X_train.columns if X_train[col].dtype == "object"]
object_cols
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_train[object_cols] = ordinal_encoder.fit_transform(X_train[object_cols])
X_test[object_cols] = ordinal_encoder.transform(X_test[object_cols])
```

```python

reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_predict=reg.predict(X_test)

```

```python
y_predict
```

```python
mae=mean_absolute_error(y_predict,y_test)
```

```python

print('variance score: {}'.format(reg.score(X_test,y_test)))
```

```python
print("Mean absolute error is:- ",mae)
```
