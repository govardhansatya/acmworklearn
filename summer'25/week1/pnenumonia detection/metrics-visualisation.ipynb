```python
import matplotlib.pyplot as plt
epochs = list(range(1, 11))
loss = [0.1759, 0.0872, 0.0639, 0.0552, 0.0367, 0.0417, 0.0236, 0.0115, 0.0203, 0.0215]
accuracy = [0.9346, 0.9670, 0.9772, 0.9797, 0.9870, 0.9843, 0.9929, 0.9964, 0.9921, 0.9919]

max_loss = max(loss)
max_acc = 1.0 

# Plot
plt.figure(figsize=(8, 5))
plt.plot(epochs, loss, 'r-o', label='Loss')
plt.plot(epochs, accuracy, 'g-s', label='Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Training Loss and Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-darkgrid')  
epochs = np.arange(1, 11)
loss = np.array([0.1759, 0.0872, 0.0639, 0.0552, 0.0367, 0.0417, 0.0236, 0.0115, 0.0203, 0.0215])
accuracy = np.array([0.9346, 0.9670, 0.9772, 0.9797, 0.9870, 0.9843, 0.9929, 0.9964, 0.9921, 0.9919])

# Normalize both to [0, 1] for shared Y-axis
loss_norm = (loss - loss.min()) / (loss.max() - loss.min())
accuracy_norm = (accuracy - accuracy.min()) / (accuracy.max() - accuracy.min())
plt.figure(figsize=(10, 6))
plt.plot(epochs, loss_norm, 'o-', color='#FF6B6B', label='Loss (normalized)', linewidth=2, markersize=6)
plt.plot(epochs, accuracy_norm, 's--', color='#1DD1A1', label='Accuracy (normalized)', linewidth=2, markersize=6)
plt.title('Training Loss and Accuracy (Normalized)', fontsize=16, weight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Normalized Value', fontsize=12)
plt.xticks(epochs)
plt.legend(loc='lower right', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()

```
