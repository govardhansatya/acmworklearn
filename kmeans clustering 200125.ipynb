```python
kmeans_kwargs = {"init": "random","n_init": 10,"max_iter": 300,"random_state": 42,
}
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
max_clusters = 30
inertias = []
silhouette_scores = []
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
for k in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, n_init=3, random_state=42)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
    score = silhouette_score(df, kmeans.labels_)
    silhouette_scores.append(score)
```

```python


# Create x-axis for plotting
k_values = list(range(2, max_clusters + 1))

# Ensure inertias and silhouette_scores have the same length as k_values
inertias = inertias[:len(k_values)]
silhouette_scores = silhouette_scores[:len(k_values)]

# Plot elbow curve
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(k_values, inertias, marker='o')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method')

# Second subplot - Silhouette Scores
ax2.plot(k_values, silhouette_scores, marker='o')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score vs k')

plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6, init = 'k-means++', max_iter = 300, n_init=40, random_state=0)
y_kmeans = kmeans.fit_predict(df)
```

```python
# Add cluster labels to the dataframe
df['Cluster'] = y_kmeans

print("\nCluster analysis:")
for cluster in range(6):
    print(f"\nCluster {cluster} stats:")
    cluster_data = df[df['Cluster'] == cluster]
    
    print(f"number of ships: {len(cluster_data)}")
    
    # Calculate mean values for key metrics
    print("\nmean values:")
    for col in cluster_data.columns:
        print(f"{col}: {cluster_data[col].mean():.2f}")
```
